```python
new_df = pd.get_dummies(data=new_df)
new_df.tail()
```

```python
age_df=pd.DataFrame(new_df.assign(Age=lambda x: x.Rings + 1.5))

age_df
```

```python
age_df.describe()
```

```python
from sklearn.model_selection import train_test_split
import sklearn.linear_model as sk

model = sk.LogisticRegression(solver='lbfgs', max_iter=13000)
train, test = train_test_split(age_df, test_size=0.3) 
print(train.shape)
print(test.shape)
```

```python
train_x = train.drop([ 'Rings', 'Sex_I', 'Sex_F', 'Sex_M'], axis=1)
train_y = train['Rings']

test_x = train.drop([ 'Rings','Sex_I', 'Sex_F', 'Sex_M'], axis=1)
test_y = train['Rings']



train_x.head()


test_x.head()

model.fit(train_x,train_y)

pred_y= model.predict(test_x)

pred_y
```

```python
labels_dict = {
    "predicted" : list(pred_y),
    "ground truth": test_y
}

```

```python
from numpy.ma.core import negative
labels_df = pd.DataFrame(labels_dict)

labels_df
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test_y,pred_y)
```

```python
from sklearn.metrics import recall_score
recall_score(test_y, pred_y, average='weighted')
```

```python
from sklearn.metrics import precision_score

precision_score(test_y, pred_y, average='micro')
```

```python
from sklearn.metrics import f1_score
f1_score(test_y, pred_y, average='weighted')
```
